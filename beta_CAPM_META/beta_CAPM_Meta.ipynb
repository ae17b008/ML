```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from datetime import datetime as dt, timedelta as td
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import pandas_datareader as web
```

```python
data = {}
data['S&P'] = pd.read_csv('S&P500.csv')
data['META'] = pd.read_csv('META.csv')
```

```python
data['META']
```

```python
df = {}
df1= {}
for name,info in data.items():
    df[name] = data[name].reset_index()
for name,info in df.items():
    df1[name] = df[name]['Close/Last']
final_df = pd.DataFrame(df1)    
```

```python
final_df
```

```python
final_df['META']=final_df['META'].str[1:].astype(float)
# reversing data order as the oldest one should be on top row
final_df=final_df[::-1]
```

```python
final_df
```

```python
daily_return = np.log(final_df.pct_change() + 1).dropna()
avg_daily_return=np.array(daily_return.mean())
avg_annual_return=avg_daily_return*252
cov_daily_return = daily_return.cov()
cov_annual_return=cov_daily_return*252
```

```python
print(avg_annual_return*100)
```

```python
print(cov_annual_return)
```

```python
market_cov = cov_annual_return.iloc[0,1]
var_market = cov_annual_return.iloc[0,0]
beta_meta = market_cov / var_market
print(f'Beta (5Y period) of META is : {beta_meta}')
```

# CAPM
E[ri]=rf+beta(market premium)
Assuming market premium to be 10% and risk free return to be 2.5%

```python
print("Expected annual return for Meta is :", 2.5+(10*beta_meta))
```
